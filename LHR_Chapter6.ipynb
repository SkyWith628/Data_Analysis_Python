[코드 LHR 6- 15]

```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터를 데이터프레임 형태로 읽어오기
df = pd.read_csv('temp_LHR.csv', encoding='cp949')
df.head(3)
```

[코드 LHR 6- 16]

```python
df.info()
```

[코드 LHR 6- 17]

```python
df2 = df.fillna(method='ffill')
df2.info()
```

[코드 LHR 6- 18]

```python
df2.rename(columns={'평균기온(°C)':'avg_temp'}, inplace=True)
df2.rename(columns={'최고기온(°C)':'max_temp'}, inplace=True)
df2.info()
```

[코드 LHR 6- 19]

```python
df2['일시'] = pd.to_datetime(df2['일시'], format='%Y-%m-%d')
df2.info()
```

[코드 LHR 6- 20]

```python
df3 = df2.set_index('일시')

df3.head(3)
```

[코드 LHR 6- 21]

```python
df_Sep = df3[pd.DatetimeIndex(df3.index).month == 9]

df_Sep
```

[코드 LHR 6- 22]

```python
plt.title('Seoul 2024. 9')
plt.plot(range(0, 30), df_Sep['max_temp'], label='max', c='r')
plt.plot(range(0, 30), df_Sep['avg_temp'], label='avg', c='y')

plt.xlim(1, 30)
plt.ylim(20, 37)
plt.legend()
plt.rcParams['figure.figsize'] = (8, 8)   # 그림의 가로 세로 사이즈를 설정
plt.show()
```
