```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

몸무게 = [30, 35, 40]
팔길이 = [45, 50, 55]

데이터 = np.array([몸무게, 팔길이]).T

scaler = MinMaxScaler()
정규화_데이터 = scaler.fit_transform(데이터)

print("=== 원본 데이터 ===")
print(데이터)

print("\n=== 정규화된 데이터 ===")
print(정규화_데이터)
```

[LHR 코드 11-3]

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression

키 = [155, 160, 165, 170, 175]
신발 = [230, 240, 250, 260, 270]

원본_데이터 = np.array([키, 신발]).T

print("=== 원본 데이터 ===")
print("   키   신발크기")
print(원본_데이터)
```

[LHR 코드 11-4]

```python
pca = PCA(n_components=2)
pca.fit(원본_데이터)
변환된_데이터 = pca.transform(원본_데이터)

print("\n=== 변환된 데이터 ===")
print("   제1주성분    제2주성분")
print(np.round(변환된_데이터, 2))
```

[LHR 코드 11-5]

```python
설명력 = pca.explained_variance_ratio_

print("\n=== PCA 설명력 ===")
print(f"제1주성분의 설명력: {설명력[0]:.2%}")
print(f"제2주성분의 설명력: {설명력[1]:.2%}")
```

[LHR 코드 11-9]

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression

수학 = [50, 65, 80, 90, 100]
과학 = [45, 60, 75, 85, 95]
영어 = [40, 55, 70, 80, 90]

원본_데이터 = np.array([수학, 과학, 영어]).T

print("\n=== 원본 데이터 ===")
print("   수학  과학  영어")
print(원본_데이터)
```

[LHR 코드 11-10]

```python
pca = PCA(n_components=3)
변환된_데이터 = pca.fit_transform(원본_데이터)

print("\n=== 변환된 데이터 ===")
print("   제1주성분   제2주성분   제3주성분")
print(np.round(변환된_데이터, 2))
```

[LHR 코드 11-11]

```python
설명력 = pca.explained_variance_ratio_

print("\n=== PCA 설명력 ===")
print(f"제1주성분의 설명력: {설명력[0]:.2%}")
print(f"제2주성분의 설명력: {설명력[1]:.2%}")
print(f"제3주성분의 설명력: {설명력[2]:.2%}")
```

[LHR 코드 11-12]

```python
X_pca = 변환된_데이터[:, 0].reshape(-1, 1)
Y = np.array(영어)

linear_model = LinearRegression()
linear_model.fit(X_pca, Y)

기울기 = linear_model.coef_[0]
절편 = linear_model.intercept_

print("\n=== PCA 기반 선형 회귀 분석 결과 ===")
print(f"회귀식: 영어점수 = {기울기:.2f} * 제1주성분 + {절편:.2f}")
```

[LHR 코드 11-13]

```python
새로운_데이터 = np.array([[40, 35, 30]])
새로운_데이터_pca = pca.transform(새로운_데이터)[:, 0]

예측값 = linear_model.predict(새로운_데이터_pca.reshape(-1, 1))

print(f"\n수학,과학,영어=40,35,30에 해당하는 제1주성분 값으로 예측한 영어점수: {예측값[0]:.2f}점")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

df = pd.read_csv('LHR_titanic_nomissing.csv')
numeric_features = df[['Age', 'Fare', 'SibSp', 'Parch']]

print("=== 원본 데이터 ===")
print(numeric_features)
```

[LHR 코드 11-16]

```python
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(numeric_features)

print("=== 정규화된 데이터 ===")
print(normalized_features)
```

[LHR 코드 11-17]

```python
pca = PCA()
pca_features = pca.fit_transform(normalized_features)

print("\n=== 변환된 데이터 (4차원) ===")
print("1주성분  2주성분  3주성분  4주성분")
print(np.round(pca_features, 4))
```

[LHR 코드 11-18]

```python
설명력 = pca.explained_variance_ratio_

print("\n=== PCA 설명력 ===")
print(f"제1주성분의 설명력: {설명력[0]:.2%}")
print(f"제2주성분의 설명력: {설명력[1]:.2%}")
print(f"제3주성분의 설명력: {설명력[2]:.2%}")
print(f"제4주성분의 설명력: {설명력[3]:.2%}")
```
