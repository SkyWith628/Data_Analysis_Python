```python
import pandas as pd
from sklearn.linear_model import LinearRegression

키 = [160, 165, 170, 175, 180]
몸무게 = [55, 60, 65, 70, 75]

df = pd.DataFrame({
    '키': 키,
    '몸무게': 몸무게
})

X = df[['키']]
y = df['몸무게']
df
```

[LHR 10-3]

```python
model = LinearRegression()
model.fit(X, y)

기울기 = model.coef_[0]
절편 = model.intercept_

print(f"회귀식: 몸무게 = {기울기:.2f} × 키 + {절편:.2f}")
```

[LHR 10-4]

```python
예측키 = [[190]]
예측몸무게 = model.predict(예측키)
print(f"키가 190cm일 때 예상 몸무게: {예측몸무게[0]:.1f}kg")
```

[LHR 10-5]

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumGothic'
plt.scatter(키, 몸무게, label='실제 데이터')
plt.plot(키, model.predict(pd.DataFrame(키)), 'r-', label='회귀선')
plt.xlabel('키(cm)')
plt.ylabel('몸무게(kg)')
plt.legend()
plt.show()
```

[LHR 10-6]

```python
결정계수 = model.score(X, y)
print("=== 모델 분석 결과 ===")
print(f"결정계수(R²): {결정계수:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df3 = pd.read_csv('LHR_weekly_smartphone_sleep_hours.csv', encoding='UTF-8')
X3 = df3[['Weekly_Smartphone_Hours']]
y_Sleep = df3['Weekly_Sleep_Hours']
df3
```

[LHR 10-17]

```python
modelSleep = LinearRegression()
modelSleep.fit(X3, y_Sleep)

slope_Sleep = modelSleep.coef_[0]
Intercept_Sleep = modelSleep.intercept_
print(f"회귀식: 수면시간 = {slope_Sleep:.2f} × 스마트폰사용량 + {Intercept_Sleep:.2f}")
```

[LHR 10-18]

```python
사용량 = [[35]]
Y_pred_Sleep = modelSleep.predict(사용량)
print(f"35시간의 사용량에 대한 수면시간: {Y_pred_Sleep[0]:.1f}시간")
```

[LHR 10-19]

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumGothic'
plt.scatter(X3, y_Sleep, label='수면시간데이터')
plt.plot(X3, modelSleep.predict(pd.DataFrame(X3)), 'r-', label='회귀선')
plt.xlabel('스마트폰사용시간')
plt.ylabel('수면시간')
plt.legend()
plt.show()
```

[LHR 10-20]

```python
# 10-20
결정계수_Sleep = modelSleep.score(X3, y_Sleep)
print("=== 수면시간 예측 모델 분석 결과 ===")
print(f"결정계수(R²): {결정계수_Sleep:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df4 = pd.read_csv('LHR_localpopulation.csv', encoding='UTF-8')
X4 = df4[['YearMonth']]
y_jeonbuk = df4['전라북도']
y_jeonnam = df4['전라남도']
df4
```

[LHR 10-23]

```python
modeljeonbuk = LinearRegression()
modeljeonbuk.fit(X4, y_jeonbuk)
slope_N = modeljeonbuk.coef_[0]
Intercept_N = modeljeonbuk.intercept_
print(f"회귀식: 전라북도인구 = {slope_N:.2f} × 년월 + {Intercept_N:.2f}")

modeljeonnam = LinearRegression()
modeljeonnam.fit(X4, y_jeonnam)
slope_S = modeljeonnam.coef_[0]
Intercept_S = modeljeonnam.intercept_
print(f"회귀식: 전라남도인구 = {slope_S:.2f} × 년월 + {Intercept_S:.2f}")
```

[LHR 10-24]

```python
예측년월 = [[202512]]
Y_pred_N = modeljeonbuk.predict(예측년월)
print(f"2025년 12월의 전라북도의 인구: {Y_pred_N[0]:.1f}명")

Y_pred_S = modeljeonnam.predict(예측년월)
print(f"2025년 12월의 전라남도의 인구: {Y_pred_S[0]:.1f}명")
```

[LHR 10-25]

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumGothic'
plt.plot(X4, y_jeonbuk, label='전라북도 실제 데이터')
plt.plot(X4, modeljeonbuk.predict(pd.DataFrame(X4)), 'r-', label='회귀선')
plt.xlabel('년월')
plt.ylabel('전라북도인구')
plt.legend()
plt.show()
```

[LHR 10-26]

```python
plt.rcParams['font.family'] = 'NanumGothic'
plt.plot(X4, y_jeonnam, label='전라남도 실제 데이터')
plt.plot(X4, modeljeonnam.predict(pd.DataFrame(X4)), 'r-', label='회귀선')
plt.xlabel('년월')
plt.ylabel('전라남도인구')
plt.legend()
plt.show()
```

[LHR 10-27]

```python
결정계수_N = modeljeonbuk.score(X4, y_jeonbuk)
print("=== 전라북도 인구 예측 모델 분석 결과 ===")
print(f"전라북도결정계수(R²): {결정계수_N:.4f}")

결정계수_S = modeljeonnam.score(X4, y_jeonnam)
print("=== 전라남도 인구 예측 모델 분석 결과 ===")
print(f"전라남도결정계수(R²): {결정계수_S:.4f}")
```
